```python
import pandas as pd
import matplotlib.pyplot as plt

# import the file
data = pd.read_csv('[P2-7] total test at day32 -Resin_01_PEIS.mpt', header = 58, sep = "\t", encoding ='latin1')

# column and row filtering & export
data = data[data['-Im(Z)/Ohm'] > 0]
data = data[['freq/Hz', 'Re(Z)/Ohm', '-Im(Z)/Ohm']]
data.to_csv('EIS_filtered.csv', index = False, header = False)

# plot
plt.plot(data['Re(Z)/Ohm'], data['-Im(Z)/Ohm'])
plt.xlabel('Z_RE')
plt.ylabel('-Z_IM')
```
